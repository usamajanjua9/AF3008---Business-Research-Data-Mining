# **1. Introduction to Research Design**

---

## **1.1 Meaning and Purpose of Research Design**

### üîé What is Research Design?

**Research design** is the overall plan or structure that guides how a research study will be conducted.

It answers key questions such as:

* What type of data will be collected?
* From whom will data be collected?
* How will the data be analyzed?
* How will the research question be answered?

---

### üìå Simple Definition

> Research design is the blueprint that connects research questions to data collection and analysis.

---

### üéØ Purpose of Research Design

The purpose of research design is to:

1. Provide structure to the study
2. Ensure systematic data collection
3. Avoid confusion and bias
4. Ensure research findings are reliable and valid
5. Help answer the research question logically

Without research design, research becomes random and unstructured.

---

### üè¶ Business Example (FinTech Context)

Suppose PaySecure wants to know:

> ‚ÄúWhy are students not actively using the digital wallet?‚Äù

If no research design is chosen:

* They may randomly ask a few students.
* They may only look at dashboard data.
* They may rely on assumptions.

With proper research design:

* They decide whether to use surveys, interviews, or experiments.
* They define variables (trust, security perception).
* They plan how to analyze results.

This makes the study scientific and credible.

---

## **1.2 Research Design as a Blueprint for Investigation**

### üèó Why is Research Design Called a Blueprint?

Just like an architect creates a blueprint before building a house, a researcher creates a research design before conducting the study.

Without a blueprint:

* Construction becomes chaotic.
* Errors increase.
* Resources are wasted.

Similarly, without research design:

* Data may not answer the research question.
* Wrong methods may be used.
* Results may be invalid.

---

### üìä Components of the Research Blueprint

A research design typically includes:

* Research objective
* Research questions
* Type of research (exploratory, descriptive, causal)
* Research approach (quantitative/qualitative)
* Data collection method
* Sampling method
* Analysis technique

---

### üéì Example

Research Question:
Does perceived security affect digital wallet usage?

Blueprint might include:

* Design: Causal
* Approach: Quantitative
* Method: Survey
* Analysis: Regression analysis

That structured plan is the research design.

---

## **1.3 Relationship Between Research Problem, Questions, and Design**

Research follows a logical flow:

### Step 1: Identify Research Problem

Example: Low transaction usage in PaySecure.

### Step 2: Develop Research Question

Example: Does perceived trust influence wallet usage?

### Step 3: Select Research Design

If we want to test cause-effect ‚Üí Causal design
If we want to explore unknown issues ‚Üí Exploratory design
If we want to describe patterns ‚Üí Descriptive design

---

### üîÑ Logical Connection

Research Problem ‚Üí Research Questions ‚Üí Research Design ‚Üí Data Collection ‚Üí Analysis ‚Üí Conclusions

If research design does not match the research question, the study becomes weak.

---

### ‚ö† Example of Mismatch

Research Question:
Does marketing campaign A increase usage compared to campaign B?

If student selects exploratory design instead of causal design ‚Üí mismatch.

Therefore, alignment is critical.

---

## **1.4 Role of Research Design in Ensuring Validity and Reliability**

### üìå What is Validity?

Validity means:

> Are we measuring what we claim to measure?

Example:
If you want to measure trust, but your survey only asks about app design, the research lacks validity.

---

### üìå What is Reliability?

Reliability means:

> Would we get the same results if we repeat the study?

If results change randomly every time, the study lacks reliability.

---

### üõ° How Research Design Improves Validity

* Clear variable definitions
* Proper measurement tools
* Logical structure
* Correct data sources

---

### üîÅ How Research Design Improves Reliability

* Standardized questionnaires
* Clear sampling strategy
* Consistent data collection procedures

---

### üíº Business Impact

Poor research design can lead to:

* Wrong managerial decisions
* Financial losses
* Poor strategy implementation

Strong research design leads to:

* Evidence-based decisions
* Reduced risk
* Strategic advantage

---

# ‚úÖ Key Takeaways

* Research design is the foundation of any research study.
* It connects research questions to data.
* It ensures scientific rigor.
* It prevents bias and confusion.
* It directly affects decision-making quality in business and FinTech.

# **2.1 Exploratory Research Design**

---

## üîé What is Exploratory Research Design?

### üìå Definition

**Exploratory research design** is used when the research problem is **not clearly defined** and the researcher wants to gain deeper understanding, insights, or ideas about the issue.

It is often the **first step** in research.

---

### üß† Simple Explanation

> Exploratory research helps you explore and understand a problem before trying to measure or test it.

It answers questions like:

* What is happening?
* Why might this be happening?
* What possible factors are involved?

---

## üìå Nature and Characteristics of Exploratory Research

Exploratory research has the following characteristics:

### 1Ô∏è‚É£ Flexible Structure

It does not follow strict procedures. The researcher can adjust questions or direction as new insights emerge.

### 2Ô∏è‚É£ Open-Ended

Focuses on generating ideas rather than testing hypotheses.

### 3Ô∏è‚É£ Qualitative Orientation

Often uses non-numerical data such as opinions, experiences, and perceptions.

### 4Ô∏è‚É£ Problem Clarification

Helps define the research problem more clearly.

### 5Ô∏è‚É£ Small Sample Size

Usually conducted with fewer participants.

---

### üìä Example (FinTech Context)

PaySecure notices low transaction usage.

They do not yet know:

* Whether it is due to trust.
* Lack of awareness.
* Poor app usability.
* Social influence.

So they conduct interviews with students to explore possible reasons.

At this stage, they are not testing anything ‚Äî they are understanding the problem.

---

## üìå When is Exploratory Design Appropriate?

Exploratory research is suitable when:

* The problem is new or unclear.
* There is limited existing research.
* The organization wants initial insights.
* The research aims to generate hypotheses.
* The company wants to understand user perceptions.

---

### üè¶ Business Example

A startup launches a new blockchain-based wallet.

Adoption is low.

Since no prior local research exists, the company first conducts exploratory research to understand:

* User awareness
* Perceived risks
* Cultural barriers

Only after exploration can they design a structured survey.

---

## üìå Flexible and Open-Ended Structure

Unlike descriptive or causal designs, exploratory research:

* Does not start with strict hypotheses.
* Allows questions to evolve.
* Allows new variables to emerge during discussion.

---

### üîÑ Example of Flexibility

During interviews, students unexpectedly mention:

* ‚ÄúParents do not trust digital payments.‚Äù

This new insight becomes an important research direction that was not initially considered.

Exploratory design allows such discoveries.

---

## üìå Common Exploratory Methods

Exploratory research typically uses qualitative methods.

---

### 1Ô∏è‚É£ Interviews

One-on-one conversations to gather detailed insights.

**Example:**
Interview 15 university students about their perception of digital wallets.

Advantages:

* Deep understanding
* Rich explanations

---

### 2Ô∏è‚É£ Focus Groups

Group discussions (6‚Äì10 participants) guided by a moderator.

**Example:**
Discuss security concerns in a student focus group.

Advantages:

* Diverse opinions
* Interactive discussion
* Reveals social influences

---

### 3Ô∏è‚É£ Pilot Studies

Small preliminary studies conducted before full research.

**Purpose:**

* Test survey questions
* Identify unclear questions
* Refine research instruments

---

### 4Ô∏è‚É£ Literature Exploration

Reviewing existing studies to identify:

* Gaps
* Variables
* Theories

This helps shape further research design.

---

## üìå Strengths of Exploratory Research

* Helps clarify vague problems
* Generates new ideas
* Identifies key variables
* Cost-effective in early stages

---

## üìå Limitations of Exploratory Research

* Cannot establish cause-and-effect
* Findings may not be generalizable
* Usually lacks statistical testing

---

# üéØ Summary

Exploratory research design:

* Is used when the problem is unclear
* Is flexible and open-ended
* Helps generate ideas and hypotheses
* Uses interviews, focus groups, and pilot studies
* Is often the first stage of research

# **2.2 Descriptive Research Design**

---

## üìå What is Descriptive Research Design?

### üîé Definition

**Descriptive research design** is used to describe characteristics, behaviors, patterns, or relationships within a population or phenomenon.

It answers the question:

> ‚ÄúWhat is happening?‚Äù

It does **not** explain cause-and-effect relationships.
It simply measures and describes.

---

### üß† Simple Explanation

> Descriptive research tells us ‚Äúwhat exists‚Äù and ‚Äúhow much,‚Äù but not ‚Äúwhy.‚Äù

---

## üéØ Purpose of Descriptive Research

The main purposes are:

* To measure the frequency of behaviors
* To describe characteristics of a population
* To examine patterns and trends
* To estimate proportions or averages

---

### üè¶ FinTech Example

PaySecure wants to know:

* What percentage of students use digital wallets?
* How frequently do they transact?
* What is the average monthly transaction amount?
* Which age group uses the app most?

These are descriptive questions.

---

## üìå Characteristics of Descriptive Research

### 1Ô∏è‚É£ Structured and Systematic

Descriptive research follows a clear and organized structure:

* Predefined research questions
* Predefined variables
* Standardized data collection tools

---

### 2Ô∏è‚É£ Large Sample Size

Usually involves larger samples than exploratory research to represent a population accurately.

---

### 3Ô∏è‚É£ Quantitative Orientation

Often involves numerical data such as:

* Percentages
* Means
* Frequencies
* Statistical summaries

---

### 4Ô∏è‚É£ Predefined Variables

Variables are identified before data collection.

Example:

* Independent variable: Age group
* Measured variable: Transaction frequency

---

## üìä Structured Data Collection

In descriptive research:

* Data collection tools are fixed and standardized.
* All participants answer the same questions.
* Measurement scales are defined in advance.

---

### üìã Example (Survey Structure)

Students are asked:

1. How often do you use PaySecure?

   * Daily
   * Weekly
   * Monthly
   * Rarely

2. Rate your trust in PaySecure (1‚Äì5 scale).

This structured format allows easy analysis.

---

## üìè Measurement of Variables

Descriptive research focuses on measuring variables clearly.

### Types of Variables Measured:

* Demographic variables (age, gender, income)
* Behavioral variables (usage frequency)
* Perception variables (trust, satisfaction)
* Financial variables (transaction value)

---

### Example

Variable: Transaction Frequency
Measurement: Number of transactions per month

Variable: Trust Level
Measurement: 5-point Likert scale

---

### üìå Important

Descriptive research measures relationships but does not prove causation.

For example:

You may find:

* Students with higher trust use the wallet more.

But you cannot conclude:

* Trust causes higher usage (that requires causal design).

---

## üìå Survey-Based Descriptive Design

### üîé What is Survey Research?

Survey research collects data using structured questionnaires.

It is the most common descriptive method in business research.

---

### Advantages

* Large sample coverage
* Cost-effective
* Easy statistical analysis

---

### Example (PaySecure)

Conduct survey of 300 students to measure:

* Usage frequency
* Trust perception
* Awareness level

Then calculate:

* Average trust score
* Percentage of active users

---

## üìå Observational Descriptive Design

### üîé What is Observational Research?

Researcher observes behavior without manipulating variables.

---

### Example

PaySecure analyzes app usage logs to observe:

* Average session duration
* Time spent per feature
* Drop-off points

This is descriptive because it reports patterns.

---

## üìä Difference Between Survey and Observational Design

| Survey                      | Observational                         |
| --------------------------- | ------------------------------------- |
| Collects self-reported data | Observes actual behavior              |
| Perception-based            | Behavior-based                        |
| Uses questionnaires         | Uses system logs or field observation |

---

## üìå Strengths of Descriptive Research

* Provides clear statistical summary
* Helps understand patterns
* Useful for business decision-making
* Cost-effective

---

## üìå Limitations of Descriptive Research

* Cannot establish cause-effect
* May suffer from response bias
* Depends on quality of measurement tools

---

# üéØ Summary

Descriptive research design:

* Describes ‚Äúwhat is happening‚Äù
* Uses structured data collection
* Measures predefined variables
* Commonly uses surveys and observational methods
* Does not establish causation

# **2.3 Causal Research Design**

---

## üìå What is Causal Research Design?

### üîé Definition

**Causal research design** is used to determine whether one variable **causes** changes in another variable.

It answers the question:

> ‚ÄúWhy is this happening?‚Äù
> ‚ÄúDoes X cause Y?‚Äù

---

### üß† Simple Explanation

> Causal research tries to prove a cause-and-effect relationship between variables.

---

## üìå Cause-and-Effect Relationships

A cause-and-effect relationship exists when:

* A change in one variable (cause)
* Produces a change in another variable (effect)

---

### Example (FinTech Context)

Cause: Increased perceived security
Effect: Higher digital wallet usage

If security improves and usage increases because of it ‚Üí causal relationship.

---

### Important Conditions for Causality

To prove causation, three conditions must be satisfied:

1Ô∏è‚É£ **Correlation** ‚Äì Variables must be related
2Ô∏è‚É£ **Time Order** ‚Äì Cause must occur before effect
3Ô∏è‚É£ **No Alternative Explanation** ‚Äì Other variables must be controlled

If these conditions are not met, we cannot claim causation.

---

## üìå Hypothesis Testing

Causal research always involves **hypothesis testing**.

---

### What is a Hypothesis?

A hypothesis is a **testable prediction** about the relationship between variables.

---

### Example

H‚ÇÅ: Perceived security positively influences digital wallet usage.
H‚ÇÄ: Perceived security does not influence digital wallet usage.

---

Causal research collects data and uses statistical methods to:

* Accept or reject hypotheses
* Determine strength of relationships

---

## üìå Experimental Research Design

### üîé What is an Experiment?

An experiment involves:

* Manipulating one variable (independent variable)
* Observing its effect on another variable (dependent variable)

---

### Structure of an Experiment

* Experimental group (receives treatment)
* Control group (does not receive treatment)
* Comparison of results

---

### FinTech Example

PaySecure tests two versions of app security messaging:

Group A: Standard security message
Group B: Enhanced security reassurance message

Then compare transaction rates.

If Group B shows significantly higher usage ‚Üí security messaging causes increased usage.

---

## üìå Quasi-Experimental Design

### What is Quasi-Experimental Design?

Similar to experiments but lacks full random assignment.

Used when randomization is difficult.

---

### Example

PaySecure launches new security feature in one city first.

Then compares usage:

* City with feature
* City without feature

Since users were not randomly assigned, it is quasi-experimental.

---

## üìå Control and Manipulation of Variables

### üîé Independent Variable

The variable that is manipulated.

Example: Security level.

---

### üîé Dependent Variable

The variable that is measured.

Example: Transaction frequency.

---

### üîé Control Variables

Variables kept constant to avoid interference.

Example:

* Age
* Income
* Education level

If not controlled, other factors may affect results.

---

### Why Control is Important

Without control:

* We cannot confirm that X caused Y.
* Results may be misleading.

---

## üìä Example of Variable Structure

Independent Variable (IV): Perceived Security
Dependent Variable (DV): Wallet Usage
Control Variables: Age, Gender, Income

Hypothesis:
Higher perceived security leads to higher wallet usage.

---

## üìå Strengths of Causal Research

* Identifies cause-and-effect
* Supports managerial decisions
* Enables hypothesis testing
* Provides strong scientific evidence

---

## üìå Limitations of Causal Research

* More expensive
* Time-consuming
* Requires careful design
* Ethical constraints may apply

---

## üìå Comparison with Other Designs

| Design      | Purpose            | Establishes Cause? |
| ----------- | ------------------ | ------------------ |
| Exploratory | Understand problem | ‚ùå No               |
| Descriptive | Describe patterns  | ‚ùå No               |
| Causal      | Test cause-effect  | ‚úÖ Yes              |

---

# üéØ Summary

Causal research design:

* Examines cause-and-effect relationships
* Uses hypothesis testing
* Often uses experiments or quasi-experiments
* Requires manipulation and control of variables
* Provides strongest evidence for managerial decisions

# **4. From Research Questions to Research Design**

---

## **4.1 Identifying the Research Problem**

### üîé What is a Research Problem?

A **research problem** is a clearly defined issue or gap that requires investigation.

It answers:

> What is wrong?
> What needs improvement?
> What do we not understand?

---

### üìå Key Characteristics of a Good Research Problem

* Clearly defined
* Researchable
* Relevant to business decision-making
* Based on evidence or observation

---

### üè¶ Example (FinTech Context)

Observation:
PaySecure has high registrations but low active usage.

Research Problem:
There is low active transaction usage among registered users.

Notice:
The problem is broad and needs to be narrowed down into research questions.

---

## **4.2 Formulating Research Questions**

### üîé What is a Research Question?

A research question is a specific question that guides the investigation.

It translates the broad problem into something measurable.

---

### üìå Characteristics of Good Research Questions

* Clear and specific
* Focused
* Measurable
* Aligned with research objectives

---

### üè¶ Example

Broad Problem:
Low wallet usage.

Possible Research Questions:

* Does perceived security influence wallet usage?
* Does trust affect transaction frequency?
* What factors influence student adoption of digital wallets?

These are more precise than the original problem.

---

## **4.3 Aligning Research Objectives with Research Questions**

### üîé What are Research Objectives?

Research objectives describe what the study intends to achieve.

They are written as action statements.

---

### Example

Research Question:
Does perceived security influence wallet usage?

Research Objective:
To examine the impact of perceived security on digital wallet transaction frequency among university students.

---

### üìå Relationship Between Questions and Objectives

| Research Question                | Research Objective                                          |
| -------------------------------- | ----------------------------------------------------------- |
| Does trust affect usage?         | To analyze the relationship between trust and wallet usage. |
| What factors influence adoption? | To identify determinants of digital wallet adoption.        |

Objectives should directly support answering research questions.

---

## **4.4 Mapping Research Questions to Suitable Research Designs**

This is where many students make mistakes.

The type of research question determines the type of research design.

---

### üîÑ Decision Logic Table

| Type of Question                | Suitable Design |
| ------------------------------- | --------------- |
| What issues exist?              | Exploratory     |
| What percentage use the wallet? | Descriptive     |
| Does security increase usage?   | Causal          |

---

### Example

Research Question:
Why are students not using the wallet?

‚Üí Exploratory Design (interviews)

Research Question:
What percentage of students use PaySecure weekly?

‚Üí Descriptive Design (survey)

Research Question:
Does improved security messaging increase transaction frequency?

‚Üí Causal Design (experiment)

---

## **4.5 Logical Flow from Problem ‚Üí Question ‚Üí Design ‚Üí Method**

This logical sequence is essential in academic research.

---

### üìå Step-by-Step Flow

1Ô∏è‚É£ Identify the Problem

2Ô∏è‚É£ Develop Research Question

3Ô∏è‚É£ Define Research Objectives

8Ô∏è‚É£ Draw Conclusions

---

### üè¶ Complete Applied Example

Low wallet transaction activity.

‚Üì

‚Üì

‚Üì

‚Üì

‚Üì

---

## üìå Why This Logical Flow Matters

If this sequence is not followed:

---

# üéØ Key Takeaways

* Research problem is broad.

# **5. Research Approaches**

Research approach refers to the **overall strategy used to collect and analyze data**.

It determines:

* What type of data will be collected
* How it will be analyzed
* Whether hypotheses will be tested

There are three major approaches:

1. Quantitative
2. Qualitative
3. Mixed-Method

---

# **5.1 Quantitative Research Approach**

---

## üìå What is Quantitative Research?

### üîé Definition

Quantitative research involves collecting and analyzing **numerical data** to test hypotheses and measure relationships between variables.

---

### üß† Simple Explanation

> Quantitative research deals with numbers, measurements, and statistics.

---

## üìä Numerical Data Collection

In quantitative research:

* Data is structured
* Questions are standardized
* Responses are measurable

---

### Examples of Quantitative Data

* Number of transactions per month
* Trust score (1‚Äì5 scale)
* Monthly spending amount
* App usage frequency

---

### üè¶ FinTech Example

Survey 500 students and ask:

* Rate perceived security (1‚Äì5)
* How many transactions do you perform monthly?

Both answers are numeric.

---

## üìà Statistical Analysis

Quantitative research uses statistical tools such as:

* Mean and percentage
* Correlation
* Regression
* Hypothesis testing
* t-test / ANOVA

---

### Example

Test whether:

Perceived Security ‚Üí Wallet Usage

Using regression analysis.

---

## üß™ Hypothesis Testing

Quantitative research follows the hypo-deductive model:

Theory ‚Üí Hypothesis ‚Üí Data ‚Üí Statistical Testing ‚Üí Conclusion

---

### Example

H‚ÇÅ: Perceived security positively influences wallet usage.

Data is collected and statistically tested to accept or reject the hypothesis.

---

## üìå Strengths of Quantitative Research

* Objective and measurable
* Generalizable results
* Strong statistical evidence
* Useful for decision-making

---

## üìå Limitations

* May ignore deeper context
* Cannot fully explain emotions or perceptions

---

# **5.2 Qualitative Research Approach**

---

## üìå What is Qualitative Research?

### üîé Definition

Qualitative research focuses on **textual, descriptive, and narrative data** to understand perceptions, experiences, and meanings.

---

### üß† Simple Explanation

> Qualitative research deals with words, opinions, and experiences instead of numbers.

---

## üìù Textual and Narrative Data

Qualitative data includes:

* Interview transcripts
* Focus group discussions
* Open-ended survey responses
* Observations
* Social media comments

---

### üè¶ FinTech Example

Interview students and ask:

* Why do you not trust digital wallets?
* What concerns do you have about security?

**Triangulation** means using multiple perspectives, methods, or data sources to verify findings.

* Technology Acceptance Model (TAM)
* Trust Theory

## **7.3 Advantages of Combining Methods**

* What is happening?
* Why is it happening?

---

### 2Ô∏è‚É£ Reduces Bias

Example:

Combining both reduces distortion.

---

### 3Ô∏è‚É£ Stronger Managerial Recommendations

Business decisions require strong evidence.

If findings are supported by:

* Data analysis
* Interviews
* Literature

---

### 4Ô∏è‚É£ Increased Research Depth

* Behavioral data
* Emotional insights
* Contextual factors

## **7.4 Improving Validity and Credibility**

### üìå Improving Reliability

### üìå Improving Credibility

* Logical
