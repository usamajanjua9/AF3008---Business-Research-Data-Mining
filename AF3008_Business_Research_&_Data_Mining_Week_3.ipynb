## **Week 3 ‚Äì Literature Review & Academic Integrity**

### **1. Critical Literature Review üìö**

---

### **1.1 Purpose of a Critical Literature Review üéØ**

**Definition:**
A critical literature review is a **systematic and analytical evaluation of existing research**, aimed at understanding, comparing, and synthesizing prior studies related to a research topic.

**Purpose:**

* To understand existing knowledge in a field
* To identify research gaps and limitations
* To justify the need for a new study
* To establish theoretical and conceptual foundations

A literature review does **not simply summarize** studies‚Äîit evaluates them to build a strong research argument.

**Example üß†:**
In a FinTech study, reviewing existing work on digital payments helps identify whether security, usability, or trust has been under-explored.

---

### **1.2 Importance of Critical Analysis in Academic Research üîç**

**Explanation:**
Critical analysis involves **questioning and evaluating** research rather than accepting it at face value.

Critical researchers examine:

* Research objectives and assumptions
* Methodology and data quality
* Strengths and weaknesses of studies
* Consistency of findings across studies

**Why it is important:**

* Improves research quality
* Prevents biased or weak conclusions
* Helps position the new study meaningfully

**Example üìä:**
Two studies report conflicting results on mobile banking adoption. Critical analysis helps explain differences based on sample size or research context.

---

### **1.3 Descriptive vs Critical Literature Review ‚öñÔ∏è**

| **Descriptive Literature Review** | **Critical Literature Review**           |
| --------------------------------- | ---------------------------------------- |
| Summarizes what authors have done | Evaluates how and why research was done  |
| Reports findings individually     | Compares and synthesizes studies         |
| Limited analytical depth          | Strong analytical and interpretive depth |
| Answers *‚Äúwhat exists?‚Äù*          | Answers *‚Äúwhat is missing and why?‚Äù*     |

**Example üè¶:**

* **Descriptive:** ‚ÄúSeveral studies examine FinTech adoption.‚Äù
* **Critical:** ‚ÄúWhile multiple studies examine FinTech adoption, most focus on urban users, revealing a gap in rural adoption research.‚Äù

## **2. Types of Literature Sources üìñ**

Understanding different types of literature sources helps researchers **select appropriate material**, **judge credibility**, and **build a strong literature review**.

---

### **2.1 Primary Literature Sources üß™**

**Definition:**
Primary literature sources are **original research works** that present **first-hand findings**, data, or analysis created by the authors.

**Key Characteristics:**

* Contain original data or experiments
* Written by the researchers who conducted the study
* Peer-reviewed (in most cases)

**Common Examples:**

* Journal research articles
* Conference papers
* Theses and dissertations

**Role in Research:**
Primary sources form the **core evidence base** of a literature review.

**Example üè¶:**
A journal article reporting survey results on mobile banking adoption is a primary source.

---

### **2.2 Secondary Literature Sources üîç**

**Definition:**
Secondary literature sources **analyze, interpret, or synthesize** information from primary sources.

**Key Characteristics:**

* Do not present new data
* Summarize or evaluate existing research
* Provide broader understanding of a topic

**Common Examples:**

* Review articles
* Meta-analyses
* Books based on existing research

**Role in Research:**
Secondary sources help:

* Identify research trends
* Understand theoretical foundations
* Locate relevant primary studies

**Example üìö:**
A systematic review summarizing studies on blockchain adoption in finance is a secondary source.

---

### **2.3 Tertiary Literature Sources üìò**

**Definition:**
Tertiary literature sources provide **condensed or reference-based information**, often used for **background understanding**.

**Key Characteristics:**

* Highly summarized
* Not analytical
* Not usually cited in academic research

**Common Examples:**

* Textbooks
* Encyclopedias
* Handbooks
* Industry summaries

**Role in Research:**
Tertiary sources are useful for:

* Learning basic concepts
* Understanding terminology
* Early familiarization with a topic

They should **not replace primary or secondary sources** in academic writing.

**Example üß†:**
A textbook chapter explaining basic concepts of business analytics.

---

### **2.4 Comparison of Literature Source Types ‚öñÔ∏è**

| **Source Type** | **Original Data** | **Academic Use**    |
| --------------- | ----------------- | ------------------- |
| Primary         | Yes               | Core evidence       |
| Secondary       | No                | Synthesis & trends  |
| Tertiary        | No                | Background learning |

---

### ‚úÖ **Key Takeaway for Students**

* Use **primary sources** as the foundation of your literature review
* Use **secondary sources** to understand trends and gaps
* Use **tertiary sources** only for background knowledge

A strong literature review relies on **careful selection of literature types**.

## **3. Evaluation of Literature üîé**

Evaluating literature ensures that a research study is **focused, credible, and comprehensive**. This step determines the **strength of the literature review** and the overall quality of research.

---

### **3.1 Relevance of Literature to the Research Problem üéØ**

**Definition:**
Relevance refers to how **closely a research article aligns with the research problem, objectives, and context** of a study.

**How to assess relevance:**

* Does the study address a similar research question?
* Is the business domain or context appropriate (e.g., FinTech, finance, management)?
* Are the variables or concepts related to your research focus?

**Why relevance matters:**

* Prevents inclusion of unrelated studies
* Keeps the literature review focused
* Strengthens problem justification

**Example üè¶:**
For a study on *digital wallet adoption*, an article on *mobile payment security* is relevant, while a study on *manufacturing supply chains* is not.

---

### **3.2 Quality Assessment of Research Articles üèÖ**

**Definition:**
Quality assessment evaluates whether a research article is **methodologically sound, credible, and reliable**.

**Key quality indicators:**

* Published in a peer-reviewed journal or conference
* Clear research objectives and methodology
* Appropriate data collection and analysis
* Logical interpretation of results
* Proper referencing

**Why quality assessment is important:**

* Ensures trustworthiness of findings
* Avoids weak or misleading conclusions
* Enhances academic rigor

**Example üìä:**
A paper with a clearly defined sample, validated survey instrument, and proper statistical analysis is considered high quality.

---

### **3.3 Sufficiency and Coverage of Selected Literature üìö**

**Definition:**
Sufficiency refers to whether **enough relevant literature has been reviewed** to adequately support the research topic.

**Key considerations:**

* Coverage of major theories and models
* Inclusion of recent and influential studies
* Balance between classic and contemporary research

**Why sufficiency matters:**

* Prevents superficial literature reviews
* Demonstrates deep understanding of the research area
* Supports identification of research gaps

**Example üß†:**
Reviewing only three articles on FinTech adoption may be insufficient, whereas analyzing 15‚Äì20 high-quality studies provides better coverage.

---

### ‚úÖ **Key Takeaway for Students**

Effective evaluation of literature ensures that your research is:

* **Focused** (relevant)
* **Credible** (high quality)
* **Well-grounded** (sufficient coverage)

These criteria are essential for producing a **strong academic research paper**.

## **4. Journal Quality Indicators  üìä**

Journal quality indicators help researchers **assess the credibility, impact, and reputation of academic journals**. They are widely used in business schools to evaluate research quality.

**Purpose:**

* Benchmark for **high-quality business research**

**Why it matters for students:**

**Purpose:**

* Provide metrics such as Impact Factor

**Why it matters:**

* Helps researchers compare journals objectively
* Indicates how frequently journal articles are cited

**Example:**
Checking JCR to verify whether a finance journal is well-ranked before selecting articles.

---


---

### **4.3 How Students Should Use Journal Quality Indicators üéì**

* Prefer **indexed and ranked journals**

* Avoid predatory or unindexed journals

---

### ‚úÖ **Key Takeaway for Students**

Journal quality indicators help ensure that your research is:

## **5. Referencing & Academic Integrity üß≠**

---

### **5.1 Purpose of Academic Referencing üìå**

**Definition:**
Academic referencing is the practice of **acknowledging the sources** of ideas, data, and findings used in research.

**Why referencing is essential:**

* Gives credit to original authors
* Demonstrates depth of literature review
* Allows readers to verify sources
* Strengthens credibility and transparency

**What referencing shows:**

* Your research is **evidence-based**
* You can **connect your work** with existing scholarship

**Example üè¶:**
When discussing digital banking adoption, citing prior studies shows that conclusions are grounded in validated research rather than opinion.

---



## **5.2 Overview of Common Referencing Styles üìù**

Academic referencing styles define **how sources are cited within the text** and **how full references are listed at the end of a document**. Different disciplines adopt different styles to ensure **consistency, clarity, and credibility** in scholarly communication.

---

## **Why Referencing Styles Exist**

Referencing styles help to:

* Standardize academic writing
* Ensure clarity and uniformity
* Allow readers to locate sources easily
* Distinguish disciplines and research traditions

Using the **correct referencing style** is often a **strict requirement** of journals, conferences, and universities.

---

## **Major Referencing Styles Used in Academia**

---

### **1. APA (American Psychological Association)**

**Discipline Usage:**

* Business and management
* Finance
* Marketing
* Social sciences
* Behavioral studies

**Key Characteristics:**

* **Author‚Äìdate** citation style
* Emphasis on **recency of research**
* Clear focus on publication year

**In-text Citation Format:**

* *(Author, Year)*
* *(Smith, 2022)*

**Reference List Characteristics:**

* Alphabetical order by author surname
* Publication year appears immediately after author name
* Strong formatting rules for titles, italics, and capitalization

**Example (Reference List):**

> Smith, J. (2022). Digital banking adoption in emerging markets. *Journal of Financial Innovation*, 10(2), 45‚Äì60.

**Why APA is important in this course:**

* Widely accepted in **business and FinTech research**
* Easy to integrate with **Mendeley and Overleaf**

---

### **2. IEEE (Institute of Electrical and Electronics Engineers)**

**Discipline Usage:**

* Computing
* Data mining
* Artificial intelligence
* Engineering
* Information systems

**Key Characteristics:**

* **Numbered citation style**
* Compact and space-efficient
* Citations appear in the order they are used

**In-text Citation Format:**

* [1], [2], [3]

**Reference List Characteristics:**

* Ordered numerically
* No alphabetical sorting
* Highly structured technical format

**Example (In-text):**

> Digital payment systems are growing rapidly [1].

**Example (Reference List):**

> [1] J. Smith, ‚ÄúDigital payment security models,‚Äù *IEEE Access*, vol. 8, pp. 12045‚Äì12060, 2022.

**Why IEEE is important in this course:**

* Preferred for **data mining and computing-oriented studies**
* Common in **conference-style research papers**

---

### **3. Harvard Referencing Style**

**Discipline Usage:**

* Business
* Management
* Economics
* Interdisciplinary research

**Key Characteristics:**

* Author‚Äìdate style (similar to APA)
* Less rigid formatting rules than APA
* Institutional variations exist

**In-text Citation Format:**

* *(Author, Year)*
* *(Smith 2022)*

**Reference List Characteristics:**

* Alphabetical order
* Flexible punctuation and formatting
* Emphasis on readability

**Example:**

> Smith, J. 2022, *Digital banking trends*, Journal of Financial Innovation, vol. 10, no. 2, pp. 45‚Äì60.

**Academic Note:**
Harvard style is **less standardized**, so institutions often provide their own guidelines.

---

### **4. Chicago Referencing Style**

**Discipline Usage:**

* Interdisciplinary research
* Qualitative studies
* Policy research
* Humanities‚Äìbusiness crossover studies

**Two Variants:**

* **Author‚Äìdate system**
* **Notes and bibliography system**

**Key Characteristics:**

* Highly detailed
* Flexible citation options
* Often uses footnotes

**Why it is less common here:**

* More complex
* Not standard for computing or business analytics

---

# 1. APA Referencing ‚Äì Step by Step (Manual) üü¶

APA = **Author‚ÄìYear** style.
Two parts:

1. **In-text citation**
2. **Reference list entry**

---

## 1.1 APA In-Text Citations ‚Äì How to Write Manually

### A. Single author

**Step 1 ‚Äì Identify:**

* Author surname
* Year of publication

**Step 2 ‚Äì Decide style:**

* Parenthetical: at end of sentence ‚Üí *(Surname, Year)*
* Narrative: author part of sentence ‚Üí *Surname (Year)*

**Example:**

* Parenthetical:

  > Mobile banking adoption is influenced by trust and ease of use *(Khan, 2022)*.
* Narrative:

  > *Khan (2022)* found that trust significantly affects mobile banking adoption.

---

### B. Two authors

**Pattern:** *(Surname1 & Surname2, Year)*

**Example:**

* Parenthetical:

  > *(Ali & Rahman, 2021)*
* Narrative:

  > *Ali and Rahman (2021)* reported that‚Ä¶

---

### C. Three or more authors

**Pattern:** *(Surname1 et al., Year)*

**Example:**

* Parenthetical:

  > *(Ahmed et al., 2023)*
* Narrative:

  > *Ahmed et al. (2023)* argued that‚Ä¶

---

### D. Organization as author

**Pattern:** *(Organization, Year)*

> *(World Bank, 2020)*
> The *World Bank (2020)* reported that‚Ä¶

---

### E. Multiple sources in one place

**Pattern:** *(Author1, Year; Author2, Year; Author3, Year)*

> Several studies confirm this trend *(Khan, 2020; Ali & Rahman, 2021; Ahmed et al., 2023)*.

---

## 1.2 APA Reference List ‚Äì Step-by-Step Templates

### A. Journal Article (Most Important)

**General Template:**

> AuthorSurname, Initials. (Year). Title of the article in sentence case. *Title of the Journal in Title Case and Italics*, *VolumeNumber*(IssueNumber), page‚Äìpage.
> [https://doi.org/xxxx](https://doi.org/xxxx)

**Step 1 ‚Äì Collect fields:**

* Author(s): Surname, Initial(s)
* Year (in brackets)
* Article title (only first word and proper nouns capitalized)
* Journal name (Title Case, italics)
* Volume (italics), issue (in brackets)
* Page range
* DOI (if available)

**Example:**

> Khan, A. (2022). Determinants of mobile banking adoption in emerging economies. *Journal of FinTech Innovation*, *10*(2), 45‚Äì63.
> [https://doi.org/10.1234/jfi.2022.015](https://doi.org/10.1234/jfi.2022.015)

Multiple authors:

> Khan, A., Ali, S., & Rahman, M. (2023). Customer trust in digital wallets: Evidence from Pakistan. *International Journal of Digital Finance*, *5*(1), 1‚Äì20.

---

### B. Conference Paper

**Template:**

> AuthorSurname, Initials. (Year). Title of paper in sentence case. In Initials. Surname (Ed.), *Title of conference proceedings in italics* (pp. xx‚Äìxx). Publisher / Organization.

**Example:**

> Ahmed, F., & Raza, T. (2023). Blockchain-based security framework for FinTech applications. In S. Malik (Ed.), *Proceedings of the International Conference on Emerging Technologies in Finance* (pp. 120‚Äì130). IEEE.

(If no editor, just write the proceedings title and organization.)

---

### C. Book

**Template:**

> AuthorSurname, Initials. (Year). *Title of the book in italics and sentence case*. Publisher.

**Example:**

> Saunders, M., Lewis, P., & Thornhill, A. (2019). *Research methods for business students*. Pearson.

---

### D. Website / Online Page

**Template:**

> Organization/Author. (Year, Month Day). Title of the page in sentence case. *Website Name*. URL

**Example:**

> World Bank. (2021, July 15). Digital financial services in South Asia. *World Bank Blogs*. [https://www.worldbank.org/‚Ä¶](https://www.worldbank.org/‚Ä¶)

(If no date: use *(n.d.)* instead of year.)

# 2. IEEE Referencing ‚Äì Step by Step (Manual) üüß

IEEE = **Numbered** style.
Two parts:

1. **In-text** ‚Äì [1], [2], [3]
2. **Reference list** ‚Äì numbered list, in order of first appearance

---

## 2.1 IEEE In-Text Citations ‚Äì How to Write Manually

### A. Single source

**Step 1 ‚Äì Find its reference number** (from your reference list)
**Step 2 ‚Äì Insert** [number] where it is first mentioned.

> Digital payments have grown rapidly in recent years [1].

You can also place it inside the sentence:

> As shown in [1], mobile banking adoption is increasing.

---

### B. Multiple sources together

**Pattern:** [1], [3], [5] or [1]‚Äì[3] for a range

> Several models have been proposed for fraud detection [2]‚Äì[4].

---

## 2.2 IEEE Reference List ‚Äì Step-by-Step Templates

General rules:

* References are numbered: **[1], [2], [3]‚Ä¶**
* Listed in **order of appearance**, not alphabetical
* Author names: First initial(s) + surname (e.g., *A. Khan*)
* Article title in quotes, journal/proceedings in italics

---

### A. Journal Article

**Template:**

> [#] A. A. Author and B. B. Author, ‚ÄúTitle of the article in quotes,‚Äù *Title of the Journal in Italics*, vol. x, no. y, pp. xx‚Äìxx, Month Year.

**Step 1 ‚Äì Collect fields:**

* Authors ‚Üí initials + surname
* Article title ‚Üí inside double quotes
* Journal name ‚Üí italic
* Volume, issue
* Page numbers
* Month + Year (if available)

**Example:**

> [1] A. Khan and S. Ali, ‚ÄúDeterminants of mobile banking adoption in emerging economies,‚Äù *Journal of FinTech Innovation*, vol. 10, no. 2, pp. 45‚Äì63, Feb. 2022.

Multiple authors: use commas + ‚Äúand‚Äù before last.

---

### B. Conference Paper

**Template:**

> [#] A. A. Author and B. B. Author, ‚ÄúTitle of paper in quotes,‚Äù in *Title of Conference in Italics*, Location, Year, pp. xx‚Äìxx.

**Example:**

> [2] F. Ahmed and T. Raza, ‚ÄúBlockchain-based security framework for FinTech applications,‚Äù in *Proc. Int. Conf. Emerging Technologies in Finance*, Islamabad, Pakistan, 2023, pp. 120‚Äì130.

(‚ÄúProc.‚Äù for Proceedings is common.)

---

### C. Book

**Template:**

> [#] A. A. Author, *Title of the Book in Italics*. City, Country: Publisher, Year.

**Example:**

> [3] M. Saunders, P. Lewis, and A. Thornhill, *Research Methods for Business Students*. Harlow, U.K.: Pearson, 2019.

---

### D. Chapter in Edited Book

**Template:**

> [#] A. A. Author, ‚ÄúTitle of chapter in quotes,‚Äù in *Title of Book in Italics*, X. Editor and Y. Editor, Eds. City, Country: Publisher, Year, pp. xx‚Äìxx.

---

### E. Website / Online Source

**Template:**

> [#] Organization/Author, ‚ÄúTitle of page in quotes,‚Äù *Website Name*. Accessed: Month Day, Year. [Online]. Available: URL

**Example:**

> [4] World Bank, ‚ÄúDigital financial services in South Asia,‚Äù *World Bank Blogs*. Accessed: Jul. 20, 2023. [Online]. Available: [https://www.worldbank.org/‚Ä¶](https://www.worldbank.org/‚Ä¶)

# **Research & Journal Quality Indicators**

*(Definitions, Meaning, and Formulas)*

These indicators are used to **evaluate the quality, impact, and credibility** of:

* Journals
* Authors
* Conferences
* Research output

---

## **1. Impact Factor (IF)**

**Used for:** Journals
**Provided by:** Journal Citation Reports (JCR)

### **What It Means (Simple Explanation)**

Impact Factor tells us:

> *On average, how many times articles from a journal are cited in other research papers.*

It focuses only on **recent publications (last 2 years)**.

---

### **Formula**

$$
\text{Impact Factor (IF)} =
\frac{\text{Citations in Year } X \text{ to articles published in } (X-1) \text{ and } (X-2)}
{\text{Total citable articles published in } (X-1) \text{ and } (X-2)}
$$

### **How to Understand This Formula**

* **Numerator:**
  How many times papers from the last 2 years were cited
* **Denominator:**
  How many research papers the journal published in those 2 years

### **Interpretation**

* Higher IF ‚Üí journal is cited more often
* IF varies by field (finance IF ‚â† medicine IF)

‚ö†Ô∏è **Never compare IF across disciplines**

---

## **2. Five-Year Impact Factor**

### **What It Means**

Same idea as Impact Factor, but:

* Uses **5 years instead of 2**
* Gives a **more stable and reliable measure**

---

### **Formula**

$$
\text{5-Year IF} =
\frac{\text{Citations in Year } X \text{ to articles published in } (X-1) \text{ to } (X-5)}
{\text{Total citable articles published in } (X-1) \text{ to } (X-5)}
$$

### **Why It Matters**

* Less sensitive to short-term citation spikes
* Better for slow-moving disciplines like business & management

---

## **3. CiteScore**

**Used for:** Journals
**Provided by:** Scopus

### **What It Means**

CiteScore measures:

> *Average citations per document over the last 4 years.*

---

### **Formula**

$$
\text{CiteScore} =
\frac{\text{Total citations received in 4 years}}
{\text{Total documents published in 4 years}}
$$

### **Why CiteScore Is Different**

* Counts **more document types** (articles, reviews, conference papers)
* Uses a **4-year window** instead of 2

üìå Often considered **more transparent** than IF.

---

## **4. h-Index**

**Used for:** Authors, journals, institutions
**Available on:** Google Scholar, Scopus, Web of Science

### **What It Means**

An author has an **h-index = h** if:

> *They have **h papers**, and **each of those papers has at least h citations***.

---

### **Example**

If an author has:

* 10 papers cited **at least 10 times each**
  ‚Üí **h-index = 10**

### **Why It Is Important**

* Combines **quantity** (number of papers)
* With **quality** (citations)

‚ö†Ô∏è There is **no single equation**; it is computed by sorting citation counts.

---

## **5. i10-Index**

**Used only by:** Google Scholar

### **What It Means**

The **number of papers that have received at least 10 citations**.

---

### **Formula**

$$
\text{i10-index} = \text{Number of papers with } \ge 10 \text{ citations}
$$

### **Pros & Cons**

‚úÖ Easy to understand
‚ùå Too simplistic for serious evaluation
‚ùå Not used by Scopus or Web of Science

---

## **6. g-Index**

### **What It Means**

The g-index improves the h-index by giving **extra importance to highly cited papers**.

---

### **Rule Explained Simply**

Top **g papers together must have at least g¬≤ total citations**.

---

### **Formula**

$$
\sum_{i=1}^{g} \text{Citations}_i \ge g^2
$$

### **Why It Exists**

* h-index ignores *very highly cited* papers
* g-index rewards researchers with blockbuster papers

---

## **7. SNIP (Source Normalized Impact per Paper)**

**Used for:** Journals
**Provided by:** Scopus

### **What It Means**

SNIP adjusts citation counts based on **how often people cite in that field**.

---

### **Conceptual Formula**

$$
\text{SNIP} =
\frac{\text{Citations per paper}}
{\text{Citation potential of the field}}
$$

### **Why SNIP Is Important**

* Some fields cite more (medicine)
* Some cite less (business)
* SNIP makes comparison **fairer across disciplines**

---

## **8. SJR (SCImago Journal Rank)**

**Used for:** Journals
**Provided by:** SCImago / Scopus

### **What It Measures**

Journal prestige based on:

* Number of citations
* Prestige of the journals giving those citations

üìå **Not all citations are equal**.

> A citation from a top journal counts more than one from a low-ranked journal.

---

## **9. Quartile Rankings (Q1‚ÄìQ4)**

**Used by:** Scopus and Web of Science

### **What It Means**

Journals are ranked **within their subject area**:

| Quartile | Meaning          |
| -------- | ---------------- |
| Q1       | Top 25% journals |
| Q2       | Top 50%          |

## **6. Ethical Research Practices ‚öñÔ∏è**

Ethical research practices ensure that academic work is **honest, original, transparent, and trustworthy**. These practices are fundamental to maintaining the credibility of research and researchers.

---

### **6.1 Understanding Plagiarism üö´**

**Definition:**
Plagiarism is the act of **using someone else‚Äôs ideas, words, data, or work without proper acknowledgment**, whether intentionally or unintentionally.

**Common Forms of Plagiarism:**

* **Direct plagiarism:** Copying text word-for-word without citation
* **Paraphrasing plagiarism:** Rewriting ideas without crediting the source
* **Self-plagiarism:** Reusing one‚Äôs own previously submitted work without disclosure
* **Source-based plagiarism:** Incorrect or misleading citations

**Why plagiarism is serious:**

* Violates academic ethics
* Undermines research credibility
* Can lead to academic penalties

**Example üß†:**
Copying a paragraph from a journal article into an assignment without citation, even if slightly reworded, is plagiarism.

---

### **6.2 Responsible Use of Academic and AI Tools ü§ñ**

**Explanation:**
Academic tools (e.g., reference managers, statistical software) and AI tools (e.g., text summarizers, assistants) can **support research**, but must be used responsibly.

**Acceptable Use:**

* Generating ideas or outlines
* Summarizing large volumes of literature (for understanding)
* Improving language clarity and structure

**Unacceptable Use:**

* Submitting AI-generated text as original work
* Using AI to fabricate citations or data
* Replacing critical thinking and analysis with AI output

**Best Practice:**

* AI tools should **assist**, not **replace**, the researcher
* Students remain fully responsible for the final content

---

### **6.3 Maintaining Originality in Research ‚úçÔ∏è**

**Definition:**
Originality refers to producing **independent analysis, interpretation, and synthesis**, even when building on existing literature.

**How to maintain originality:**

* Use literature to support arguments, not copy them
* Synthesize findings across multiple studies
* Clearly distinguish prior work from your own contribution

**Example üìä:**
Combining results from multiple FinTech studies to propose a new conceptual framework demonstrates originality.

---

### **6.4 Transparency in Research Practices üîç**

**Explanation:**
Transparency means being **open and clear** about how research was conducted and how tools and data were used.

**Key aspects of transparency:**

* Clear description of methodology
* Honest reporting of results and limitations
* Proper disclosure of tool usage (including AI, if required)

**Why transparency matters:**

* Builds trust with readers and reviewers
* Enables reproducibility
* Strengthens academic credibility

---

### ‚úÖ **Key Takeaway for Students**

Ethical research is not just about avoiding misconduct, it is about:

* **Honesty** in scholarship
* **Responsibility** in tool usage
* **Originality** in thinking
* **Transparency** in reporting

These principles are essential for producing a **credible, publishable research paper**.
